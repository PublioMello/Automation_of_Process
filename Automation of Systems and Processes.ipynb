Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

File path of the sales from the previous day:

Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing


```python
import pyautogui
import pyperclip
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time

#create the chorme service/cria o servico pro chrome 
servico = Service(ChromeDriverManager().install())

```

```python
#wait 1 second between every command/ espera um segundo entre cada comando
pyautogui.pause = 1

#iniciate the chrome/inicializa o chorme
navegador = webdriver.Chrome(service=servico)
#opens the page that we need/ abre a pagina que precisa
navegador.get(" https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
time.sleep(5)
navegador.find_element(By.CLASS_NAME, 'KL4NAf').click()


#waits untill the element to be loaded/ espera o elemento carregar
while len(navegador.find_elements(By.CLASS_NAME, 'uXB7xe')) < 1:
    time.sleep(1)
time.sleep(1) # uarantee / garantia

navegador.find_element(By.CLASS_NAME, 'uXB7xe').click()

#waits untill the element to be loaded/ espera o elemento carregar
while len(navegador.find_elements(By.XPATH, '/html/body/div[12]/div[4]/div/div[3]/div[2]/div[2]/div[3]/div')) < 1:
    time.sleep(1)
time.sleep(1) # guarantee / garantia

#Click on the button to download the file/ clica no botão para fazer download do arquivo
navegador.find_element(By.XPATH, '/html/body/div[12]/div[4]/div/div[3]/div[2]/div[2]/div[3]/div').click()
#since its a small file 7 seconds its enough before close the page/ ja que é um arquvio pequeno 7 segundos é o suficiente antes de fechar a pagina
time.sleep(7)
#Fecha a pagina de web
navegador.close()

```

### Now We're gonna read the archive that it was downloaded
### Vamos agora ler o arquivo baixado para pegar os indicadores

- Billing / Faturamento
- Quantity of products / Quantidade de Produtos

```python
#calculating the kpi/calculando os kpis
import pandas as pd

#open and read the file as a table/ abre e le o arquivo como tabela
tabela = pd.read_excel(r'C:\Users\User\Downloads\Vendas - Dez.xlsx')
display(tabela)

#make the sum of the values that we need / faz a soma dos valores que queremos
faturamento = tabela['Valor Final'].sum()
qtd_produtos = tabela['Quantidade'].sum()
print('O faturamento da empresa foi: R${:,}'.format(faturamento))
print('A quantidade de vendas foi de {:,}'.format(qtd_produtos))
```
